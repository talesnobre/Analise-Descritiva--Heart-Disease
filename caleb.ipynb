```python
import pandas as pd
import numpy as np
import plotly.express as px
arquivo = pd.read_csv('heart.csv')
df = pd.DataFrame(arquivo, copy=True)
```

```python
# Histograma - Idade com Doença Cardíaca

fig = px.histogram(df, x="Age", color="HeartDisease", width=400, height=400, marginal="box")
fig.update_traces(marker_line_width=1,marker_line_color="black")
fig.show()
```

```python
fig = px.histogram(df, x="Sex", color="HeartDisease", width=400, height=400)
fig.update_traces(marker_line_width=1,marker_line_color="black")
fig.show()
```
